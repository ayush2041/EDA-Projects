#### country code

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country
```

## Merged both zomato and country code file

```python
final_df =pd.merge(df, df_country, on='Country Code',how='left')
```

```python
final_df.head()
```

```python
## To check Data Types
final_df.dtypes
```

```python
Country_names = final_df.Country.value_counts().index
```

```python
Country_val = final_df.Country.value_counts().values
```

```python
## Pie Chart  of Top 3 countries that use zomato
plt.pie(x=Country_val[:3], labels = Country_names[:3], autopct = '%1.2f%%')
```

Observatios: Zomato maximum records or transaction are from INDIA
And after that United States and then United Kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observations 
1. When Rating is between 4.5 to 4.9 ---> Excellent
2. When Rating is between 4.0 to 4.4 ---> very good
3. When Rating is between 3.5 to 3.9 ---> good
4. When Rating is between 2.5 to 3.4 ---> Average
5. When Rating is between 1.8 to 2.4 ---> Poor

```python

ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, hue='Rating color',palette=['lightgrey', 'red', 'orange', 'yellow', 'green', 'darkgreen'])
```

Observation: 
1. Not rated count is very High
2. Maximum number of rating is between 2.5 to 4.4 

```python
## Count plot for the Rating color
sns.countplot(x='Rating color', data=ratings, palette=['lightgrey', 'red', 'orange', 'yellow', 'green', 'darkgreen'])
```
